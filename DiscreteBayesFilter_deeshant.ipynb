```python
def discrete_bayes(bel_x, Pzx, Px_xbar):
  eta = 0
  bel_x_bar = []
  for x in range(100):
    #print(bel_x_bar)
    bel_x_bar.append(float("{:.10f}".format(Pzx[x]*bel_x[x])))
    eta = eta + bel_x_bar[x]
  bel_x_bar = [float("{:.10f}".format(bel_x_bar[x]/eta)) for x in range(100)]
  for x in range(100):
    bel_x_bar[x] = sum([float("{:.10f}".format(Px_xbar[x_bar]*bel_x_bar[x_bar])) for x_bar in range(100)])
  return bel_x_bar

```

```python
import matplotlib.pyplot as plt
bel_x = [0.01]*101
#total time steps=100+1
#print(bel_x)
for timestamp in range(50):
  plt.figure()
  plt.plot(bel_x)
  plt.title("belief after time steps")
  
  plt.xlabel("number of steps")
  Pzx = [0]*100
  Px_xbar = [0]*100
  for state in range(100):
    if timestamp == state:
      # z- measurement
      Pzx[state] = 0.9/1

    else : Pzx[state] = 0.1/100

    Px_xbar[timestamp] = 0.05
    if timestamp - 1 >= 0 : Px_xbar[timestamp - 1] = 0.9
    if timestamp + 1 < 50 : Px_xbar[timestamp + 1] = 0.05
  bel_x = discrete_bayes(bel_x, Pzx, Px_xbar)
    
```

```python
print("belief after 50 time steps")
print(bel_x)
plt.plot(bel_x)
plt.title("final belief plot after 50 time steps")
plt.xlabel("number of steps")
```
